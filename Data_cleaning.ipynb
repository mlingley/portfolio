```python
df2 = pd.read_csv(r'C:\Users\malin\Downloads\vchartz_data.csv')
df2
```

```python
df2["Total Shipped in Millions"] = df2["Total Shipped"].str.replace('m','').astype('float')
df2["Total Sales in Millions"] = df2["Total Sales"].str.replace('m','').astype('float')
df2["NA Sales in Millions"] = df2["NA Sales"].str.replace('m','').astype('float')
df2["PAL Sales in Millions"] = df2["PAL Sales"].str.replace('m','').astype('float')
df2["Japan Sales in Millions"] = df2["Japan Sales"].str.replace('m','').astype('float')
df2
```

```python
df2.to_csv(r"C:\Users\malin\Downloads\vgchartz_data_cleaned.csv", index=False)
```

```python
df1 = pd.read_csv(r'C:\Users\malin\Downloads\metacritic_games_cleaned.csv')
df2 = pd.read_csv(r'C:\Users\malin\Downloads\vgchartz_data_cleaned.csv')
```

```python
df = pd.merge(df1, df2, on='Title', how='outer')
df
```

```python
def normalize_text(text):
    return re.sub(r'[^\w\s]', '', text).lower()
```

```python
df['Title'] = df['Title'].apply(normalize_text)
df
```

```python
df.drop(columns=['User Score','Release Date','Last Update'], inplace=True)
df
```

```python
df.to_csv(r"C:\Users\malin\Downloads\metacritic_and_vgchartz_join3.csv", index=False)
```
