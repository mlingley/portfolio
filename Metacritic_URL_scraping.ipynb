```python
# Importing packages

import numpy as np
import pandas as pd
import time
import csv
from bs4 import BeautifulSoup, SoupStrainer
import lxml
import html5lib
import pprint
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import scipy as sp
from urllib.request import Request, urlopen
import re
import requests,json

pd.options.mode.chained_assignment = None
```

```python
#check if website is allowing me to scrape

headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
html_text = requests.get("https://www.metacritic.com/browse/game/?releaseYearMin=1910&releaseYearMax=2023", headers=headers)

print(html_text)
```

```python

#get url tags from the first page

url = 'https://www.metacritic.com/browse/game/?releaseYearMin=1910&releaseYearMax=2023'
response = requests.get(url, headers = headers)
soup=BeautifulSoup(response.text,'html.parser')

allLinks = []

for tag in soup.find_all(href=re.compile("^/game/")):
    urls = tag['href']
    allLinks.append(urls)
print(allLinks)
```

```python
#changing the step from above to iterate through multiple pages

base_url = 'https://www.metacritic.com/browse/game/?releaseYearMin=1910&releaseYearMax=2023&page='
allLinks = set()
current_page = 1
max_retries = 5

while current_page <=700:
  url = base_url + str(current_page)
  retries = 0

  while retries < max_retries:
    try:
      response = requests.get(url, headers = headers)
      response.raise_for_status()
      soup=BeautifulSoup(response.text,'html.parser')

      game_links = soup.find_all(href=re.compile("^/game/"))
      if not game_links:
        break
      for tag in game_links:
        link = tag['href']
        allLinks.add(link)

      current_page += 1
      break
    except requests.RequestException as e:
      print(f"Error: {e}")
      retries += 1
  else:
    print("Max retries exceeded. Skipping page.")

#print(list(allLinks))
print(f"Total games found: {len(fullurl)}")
```
